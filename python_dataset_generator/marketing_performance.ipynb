```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
```

```python
# --------------------------------------------------
# CONFIGURATION
# --------------------------------------------------

weeks = pd.date_range(start="2024-01-01", periods=104, freq="W-MON")

channels = {
    "Email": {
        "spend_range": (8000, 14000),
        "ctr": 0.08,
        "lead_rate": 0.12,
        "conversion_rate": 0.035,
        "aov": 850
    },
    "Referral": {
        "spend_range": (7000, 13000),
        "ctr": 0.07,
        "lead_rate": 0.11,
        "conversion_rate": 0.03,
        "aov": 800
    },
    "Google": {
        "spend_range": (10000, 20000),
        "ctr": 0.05,
        "lead_rate": 0.09,
        "conversion_rate": 0.025,
        "aov": 650
    },
    "Facebook": {
        "spend_range": (13000, 23000),
        "ctr": 0.04,
        "lead_rate": 0.08,
        "conversion_rate": 0.020,
        "aov": 550
    },
    "LinkedIn": {
        "spend_range": (12000, 24000),
        "ctr": 0.025,
        "lead_rate": 0.05,
        "conversion_rate": 0.018,
        "aov": 1000
    }
}

regions = ["North America", "Europe", "APAC"]
devices = ["Desktop", "Mobile"]
campaign_types = ["Awareness", "Retargeting", "Acquisition"]

# --------------------------------------------------
# DATA GENERATION
# --------------------------------------------------

rows = []

for week in weeks:

    seasonality = 1 + 0.15 * np.sin((week.weekofyear / 52) * 2 * np.pi)

    for channel, cfg in channels.items():

        spend = np.random.uniform(*cfg["spend_range"]) * seasonality

        impressions = int(spend * np.random.uniform(45, 70))

        clicks = int(impressions * cfg["ctr"])

        leads = int(clicks * cfg["lead_rate"])

        customers = int(leads * cfg["conversion_rate"])

        revenue = customers * cfg["aov"] * np.random.uniform(0.9, 1.15)

        profit_margin = np.random.uniform(0.25, 0.42)

        profit = revenue * profit_margin

        rows.append({
            "week_start": week,
            "channel": channel,
            "spend": round(spend, 2),
            "impressions": impressions,
            "clicks": clicks,
            "leads": leads,
            "customers": customers,
            "revenue": round(revenue, 2),
            "profit": round(profit, 2),
            "campaign_type": np.random.choice(campaign_types),
            "region": np.random.choice(regions),
            "device": np.random.choice(devices)
        })

# --------------------------------------------------
# CREATE DATAFRAME
# --------------------------------------------------

df = pd.DataFrame(rows)

# --------------------------------------------------
# ADD DERIVED METRICS
# --------------------------------------------------

df["cac"] = (df["spend"] / df["customers"]).round(2)

df["conversion_rate"] = (
    df["customers"] / df["leads"]
).round(4)


df["roi"] = (
    df["revenue"] / df["spend"]
).round(2)

# --------------------------------------------------
# EXPORT CSV
# --------------------------------------------------

df.to_csv("realistic_mmm_dataset.csv", index=False)

print(df.head())
print("\nDataset Shape:", df.shape)
print("\nCSV Exported Successfully")
```
